```python
# La funcion a resolver

def f(v):
    
    p=10 #atm
    T= 353.2 #kelvin
    R=0.08205 #atm x l/mol x k
    a=3.599
    b=0.04267

    value =(p*(v**3))-(p*b*(v**2))+(a*v)-(a*b)-(R*T*(v**2))
    return value
```

```python
# Inputs

p_prev_2 = 2.
p_prev_1 = 5.0
Tol = 1e-8
error = 1.0
Nmax =1000
iteracion = 0
```

```python
# Implementacion de algoritmo

while (error>Tol):
    p = p_prev_1 - (p_prev_1 - p_prev_2)*f(p_prev_1)/(f(p_prev_1)-f(p_prev_2))
    error = abs(p-p_prev_1)
    if (f(p)*f(p_prev_1)<0):
        p_prev_2 = p_prev_1
        p_prev_1 = p
        print("p_n = %.16f\t error = %.16f\t abs(f(p_n)) = %.16f" %(p,error,f(p)))
    elif (f(p)*f(p_prev_1)>0):
        p_prev_1 = p_prev_2
        p_prev_2 = p
        print("p_n = %.16f\t error = %.16f\t abs(f(p_n)) = %.16f" %(p,error,f(p)))
    
    iteraccion = iteracion+1
    if(iteracion>= Nmax):
        print("El método no converge despues de %d iteracciones" %Nmax)
        break
        
        
```
