# Cancelaci√≥n sustractiva

```python
a=1.0
b=1.0
x_list =[]
y_list =[]

for i in range (1,10):
    c=10**(-i)
    x_1= (-b+(b**2-4*a*c)**0.5)/(2*a)       
    x_1_p = 2*c/(-b-(b**2-4*a*c)**0.5)
    error = abs(x_1_p - x_1)/ abs(x_1_p)*100.
    
    x_list.append(c)
    y_list.append(error)
    print("c=%.5e x_1=%.16e x_1_p=%.16e error=%.16ef"%(c,x_1,x_1_p,error))
```

```python
##grafica del error 

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
## corrige la grafica (sale solo una)

fig=plt.figure(figsize=(10,10))
ax=plt.axes()
ax.grid()
ax.plot(x_list,y_list,'.')
ax.set_yscale('log') ##modificando a escala logaritmica
ax.set_xscale('log')
ax.set_xlabel("c")
ax.set_ylabel("error")
plt.show()
```
