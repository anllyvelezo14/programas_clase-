```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#subiendo los datos en jupyer

datos = np.loadtxt("datos.dat") ##Carga los datos
x_list=[]
y_list=[]

for i in range (0,datos.shape[0]): ##llenando los arreglos de arriba con los datos de la matriz Con los ptos expert
    x_list.append(datos[i][0])
    y_list.append(datos[i][1])
    
print(datos.shape) ## .shape retorna el tamaÃ±o de la matriz
print(x_list)
print(y_list)

def interpolacionlineal (x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if(x>=x_list[i] and x<=x_list[i+1]):
            x_prev = x_list[i] #x_i
            y_prev = y_list[i] #y_i
            x_next = x_list[i+1]  #x_i+1
            y_next = y_list[i+1]   #x_i+1
            
            y =(y_next - y_prev)/(x_next-x_prev)*(x-x_prev)+y_prev
            
            return y
           
##print(interpolacionlineal(x_list,y_list,4.1))

x_interpol = np.linspace(min(x_list),max(x_list),100)
y_interpol = []

for i in range (0,len(x_interpol)):
    y_interpol.append(interpolacionlineal(x_list,y_list,x_interpol[i]))
print(x_interpol)
```

```python
fig = plt.figure()
ax=plt.axes()
ax.plot(x_list,y_list,'*', markersize='10', label='datos discretos')
ax.plot(x_interpol,y_interpol,'.',label='puntos interpolados')
ax.plot(x_interpol,np.sin(x_interpol),label='sin(x)')
ax.legend()
plt.show()
```
