# Derivación previa

Derivación simbolica y numérica con python

```python
## Se quiere realizar la derivada analitica (simbolica) de una función, python ofrece una libreria que
## muestra la derivada analitica de una función

  #se importa la libreria
from sympy import Symbol

x = Symbol('x')                                        #se escribe la variable x como un simbolo
y = 2*x**3                                             #se define la funcion a derivar 

Derivada = y.diff(x)                                   #se define la variable derivada para 
                                                       #almacenar la funcion analitica 

print("La derivada de la función es: ",Derivada)
```

```python
## Se quiere obtiene la derivada numerica de la funcion con python
## esa funcion de la libreria usa el metodo de diferencias divididas para obtener la derivada 

  #se importa la libreria
from scipy.misc import derivative
x_0 =1                                                # x_0punto en el cual se va a evaluar la derivada

f = lambda x:2*x**3                                   # funcion f en terminos de x

y = derivative(f,x_0,0.001)                           # la funcion derivative requiere de 
                                                      # (función, pto x_0 , dx= )  donde dx es el espaciamiento
                                                      # entre los datos, es decir h. MIentras mas pequeño, mejor
                                                      # la aproximacion

print ("La derivada de la función evaluada en", x_0, "es:" ,y)
```

# Conociendo la función

 According to the formal definition of differentiation, given a function $f(x)$ such that $f(x)\in C^1[a,b]$, the first order derivative is given by
 
 
 $$f'(x) \approx \frac{f(x+h)-f(x)}{h}$$
where the function must be known at least in $x_0$ and $x_1 = x_0+h$, and $h$ should be small enough.

EJEMPLO

Evaluate the first derivative of the next function using the previous numerical scheme at the point $x_0=2.0$ and using $h=0.5,\ 0.1,\  0.05$

$f(x) = \sqrt{1+\cos^2(x)}$

```python

import numpy as np
import matplotlib.pyplot as plt


#Function to evaluate
def function(x):
    return np.sqrt( 1+np.cos(x)**2 )

#X value
x0 = 2
xmin = 1.8
xmax = 2.2
#h step
hs = [0.5,0.1,0.05]

#Calculating derivatives
dfs = []
for h in hs:
    dfs.append( (function(x0+h)-function(x0))/h )
    
#Plotting
plt.figure( figsize=(10,8) )
#X array
X = np.linspace( xmin, xmax, 100 )
Y = function(X)
plt.plot( X, Y, color="black", label="function", linewidth=3, zorder=10 )
#Slopes
Xslp = [1,x0,3]
Yslp = [0,0,0]
for df, h in zip(dfs, hs):
    #First point
    Yslp[0] = function(x0)+df*(Xslp[0]-Xslp[1])
    #Second point
    Yslp[1] = function(x0)
    #Third point
    Yslp[2] = function(x0)+df*(Xslp[2]-Xslp[1])
    #Plotting this slope
    plt.plot( Xslp, Yslp, linewidth = 2, label="slope$=%1.2f$ for $h=%1.2f$"%(df,h) )

#Format
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim( xmin, xmax )
plt.ylim( 1, 1.15 )
plt.legend( loc = "upper left" )
```

Método de los tres puntos

```python
##importando librerias
import numpy as np
import matplotlib.pyplot as plt

##definiendo las listas de los puntos discretos, deben estar ordenadas
x_list = [1.8,1.9,2.,2.1,2.2]
y_list = [10.889365,12.703199,14.778112,17.148957,19.855030]
```

```python
## Inicialmente, se deben definir los valores dependiendo de la 
## poscision del punto sobre el cual se va a evaluar la derivada

# se crea una lista para almacenar las derivadas
dydx_list = []

def dfdx(x_list, y_list):                         # Se define la funcion derivada en terminos de las
                                                  # listas de datos y el valor x donde se va a evaluar
                                                  # o se hace un ciclo for para que devuelva la derivada
                                                  # en todos los puntos (i)
    for i in range (0,len(x_list)): 
        
    # Si i es el extremo izquierdo
        if i == 0:
        
             x0 = x_list[i]
             y0 = y_list[i]
        
             x1 = x_list[i+1]
             y1 = y_list[i+1]
        
             x2 = x_list[i+2]
             y2 = y_list[i+2]
        
        
     # Si i es el extremo derecho   
        elif i ==len(x_list)-1:
        
            x0 = x_list[i]
            y0 = y_list[i]
        
            x1 = x_list[i-1]
            y1 = y_list[i-1]
        
            x2 = x_list[i-2]
            y1 = y_list[i-2]
     
        
    # Si i es un punto central
        else:
        #i= x_list.index(x)                            # .index(x) muestra la posicion de x en la lista
                                                      # ubicando a x en la lista y poder usar los ptos contiguos
            x0 = x_list[i]
            y0 = y_list[i]
        
            x1 = x_list[i-1]
            y1 = y_list[i-1]
        
            x2 = x_list[i+1]
            y2 = y_list[i+1]
        
      
        xj = x_list[i]
        
        
        dl0dx = ((2*xj-x1-x2)/(x0-x1)*(x0-x2))
        dl1dx = ((2*xj-x0-x2)/(x1-x0)*(x1-x2))
        dl2dx = ((2*xj-x0-x1/(x2-x0)*(x2-x1)))
        
        value =  dl0dx*y0 + dl1dx*y1 + dl2dx*y2    
       
        dydx_list.append(value)
    
    return dydx_list

x_array=np.linspace(min(x_list),max(x_list),100)
dfdx_array=[]

def dfdx_exact(x):
    value = np.e**x + x*np.e**x
    return value

for i in range(0,len(x_array)):
    dfdx_array.append(dfdx_exact(x_array[i]))

print(dfdx(x_list,y_list))
print(dfdx_array)   



fig=plt.figure()
ax=plt.axes()
ax.plot(x_array,dfdx_array,'-')
ax.plot(x_list,dydx_list,'.')

       
```

```python
x_list = [1.8,1.9,2.,2.1,2.2]
y_list = [10.889365,12.703199,14.778112,17.148957,19.855030]

#PRIMERA DERIVADA dydx


def dydx(x_list,y_list):
    
    #Calcular valor de h
    h=x_list[1]-x_list[0]
    
    #se crea lista vacia para almacenar las derivadas
    dydx_list=[]
    
    for i in range(0,len(x_list)):
        
        # Cuando i es el extremo derecho
        if i==0:
            
            f0 = y_list[i]
            f1 = y_list[i+1]
            f2 = y_list[i+2]
            
            value = 0.5/h*(-3*f0+4*f1-f2)
            
        # Cuando i es el extremo izquierdo
        elif i==len(x_list)-1:
            
            f0 = y_list[i-2]
            f1 = y_list[i-1]
            f2 = y_list[i]
            
            value = 0.5/h*(f0-4*f1+3*f2)
            
        # Cuando i toma un valor intermedio
        else:
            f0=y_list[i-1]
            f1=y_list[i]
            f2=y_list[i+1]
            value=0.5/h*(f2-f0)
            
        dydx_list.append(value)
        
    return dydx_list

print(dydx(x_list,y_list))

"""
SI se quiere comparar las derivadas obetnidas con el metodo con las obtenidas analiticamente
x_array=np.linspace(min(x_list),max(x_list),100)
dfdx_array=[]

def dfdx_exact(x):
    value = np.e**x + x*np.e**x
    return value

for i in range(0,len(x_array)):
    dfdx_array.append(dfdx_exact(x_array[i]))

print(dfdx(x_list,y_list))
print(dfdx_array)

"""
```

```python
#segunda derivada 

def ddyddx(x_list,y_list):
    
    #Calcular valor de h
    h=x_list[1]-x_list[0]
    
    #Creamos lista vacia para almacenar las derivadas
    ddyddx_list=[]
    
    for i in range(0,len(x_list)):
        # Usar diferencia hacia adelante
        if i==0:
            f0=y_list[i]
            f1=y_list[i+1]
            f2=y_list[i+2]
            value=1/(h*h)*(f0-2*f1+f2)
        elif i==len(x_list)-1:
            f0=y_list[i-2]
            f1=y_list[i-1]
            f2=y_list[i]
            value=1/(h*h)*(f0-2*f1+f2)
        else:
            f0=y_list[i-1]
            f1=y_list[i]
            f2=y_list[i+1]
            value=1/(h*h)*(f0-2*f1+f2)
            
        ddyddx_list.append(value)
        
    return ddyddx_list
```
